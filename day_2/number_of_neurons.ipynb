```python
# Import libraries 
import tensorflow as tf
from tensorflow import keras
from keras import layers 

# dataset 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

# Scikitlearn 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# Import the dataset 
df = sns.load_dataset('tips')

# Preprocess the data 
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['smoker']=le.fit_transform(df['smoker'])
df['day']=le.fit_transform(df['day'])
df['time']=le.fit_transform(df['time'])

# Convert 'tip' to a binary variable, 1 if the tip is above $2, else 0 
df['tip']=(df['tip'] > 2).astype(int)

# Select the features and labels variables 
X = df.drop(columns=['tip'])
y = df['tip']

# Scale the data 
scaler = MinMaxScaler()
X_scaled= scaler.fit_transform(X) 

# train test split the data 
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=[X_train.shape[1]]), # input layer ->X_train has 6 features |16 neurons here since n_neurons> n_features
    tf.keras.layers.Dense(8, activation='relu'), # hidden layer| n_neurons in hidden layer depends on the n_features
    tf.keras.layers.Dense(1, activation='sigmoid') # output layer | one neuron in output layer since we use binary classification |multiclass classification n_neurons=n_class.
])

# Compile the model
model.compile(
    optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']
)

# Fit/Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1) # batch size tells how many number of samples in one iteration has to process| verbose=1 to realize epoch running 

# Evaluate the model    
loss, accuracy= model.evaluate(X_test, y_test)
print('Accuracy:', accuracy)
print('Loss:', loss)
```
